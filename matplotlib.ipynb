<h3>(1) Line Plot</h3>
<h5>It is used to display trends or patterns over a continuous range. </h5>

```python
country_population = [750, 740, 500, 735, 657, 800, 900,]
country_population
```

```python
plt.plot(country_population)
```

```python
country_population = [750, 740, 500, 735, 657, 800, 900,]
x_axis = [50, 100, 120, 180, 200, 350, 400]
plt.plot(x_axis, country_population)
plt.show()
```

```python
country_population = [750, 740, 500, 735, 657, 800, 900,]
x_axis = [50, 100, 120, 180, 200, 350, 400]
plt.plot(x_axis, country_population, marker= '.', ms = 18, mfc = "yellow", ls = "dotted", lw= 4)
plt.title("Line PLot Example")
plt.xlabel("Axis-x")
plt.ylabel("Country Population")
plt.show()
# ms = markersizeabs
# mfc = marker face color
# ls = line style
# lw = line width
```

<h3>(2) Bar Plot</h3>

```python
country_population = [450, 730, 900, 585, 800]
country_name = ["Japan", "Bangladesh", "India", "Nepal", "China"]

# Bar Plot
plt.bar(country_name, country_population, ec = 'k', lw = 3)

plt.show()
```
